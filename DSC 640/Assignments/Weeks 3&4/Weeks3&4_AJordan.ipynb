```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import squarify as sq

%pwd
```

```python
#Import data
world = pd.read_excel('data/ex2-2/world-population.xlsm')
postage = pd.read_excel('data/ex2-2/us-postage.xlsm')
unemployment = pd.read_csv('data/ex2-2/unemployement-rate-1948-2010.csv')
expend = pd.read_csv('data/ex2-2/expenditures.txt', delimiter = "\t")
```

### Python - Tree Map

```python
#Create Tree Map of 1990 expenditures
expend_90 = expend.loc[expend['year']==1990]
sq.plot(sizes=expend_90['expenditure'],label=expend_90['category'], alpha=.8)
plt.gcf().set_size_inches(10, 8)
plt.title("Expenditure by Category")
plt.axis('off')
plt.show()
```

### Python - Area Chart

```python
#Create Area Chart of Population by Year
plt.fill_between(world['Year'], world['Population'], color='lightcoral', alpha=.3)
plt.gcf().set_size_inches(10, 8)
plt.title('Population by Year')
plt.xlabel('Year')
plt.ylabel('Population in billions')
plt.plot(world['Year'], world['Population'], color='indianred')
plt.show()
```

### Python - Stacked Area Chart

```python
#Create pivot table
expend_pivot = expend.pivot_table(index='year',columns='category',values='expenditure')
expend_pivot.head()
```

```python
#Create stacked area chart of expenditure by category by year
ax=expend_pivot.plot.area()
ax.set_xticks(expend_pivot.index)
plt.xticks(rotation=90)
plt.gcf().set_size_inches(10, 8)
plt.title('Expenditure by Category by Year')
plt.ylabel('Expenditure')
plt.xlabel('Year')
plt.show()
```

### Python - Step Chart

```python
#Create step chart of population by year
plt.step('Year','Population',data=world)
plt.title('Population by Year')
plt.xlabel('Year')
plt.ylabel('Population in billions')
plt.gcf().set_size_inches(10, 8)
plt.show()
```
