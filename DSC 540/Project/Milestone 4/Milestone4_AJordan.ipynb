```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import requests
import urllib
import pokepy #api wrapper
import pokebase as pb #api wrapper
import random
from typing import Dict, Iterable, List, KeysView, ItemsView, ValuesView
```

## 1.Create list with pokemon api data

```python
#Loop call to list
x=0
result=[]
while x < 500:
    x=x +1
    url_stat = "https://pokeapi.co/api/v2/pokemon/" + str (x) + "/" #Set url
   
    response_stat = requests.get(url_stat)
    pokelist= pd.json_normalize(json.loads(response_stat.text))[['id','name','height','weight','base_experience']].values.tolist()[0]
    result.append(pokelist)
```

## 2.Create dataframe

```python
#Create dataframe from list
api_df=pd.DataFrame(result,columns=['id','name','height','weight','base_experience'])
api_df
```

## 3.Fix Casing

```python
#capitalize first letter of each name
api_df['name'] = api_df['name'].apply(lambda x: x.capitalize())
api_df
```

```python
api_df = api_df.rename(columns={'id':'Pokedex','name':'Name'})
```
