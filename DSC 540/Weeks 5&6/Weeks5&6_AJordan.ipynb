### 1. Filter out Missing Data

```python
#Create copy of dataframe for modifications
candy17_copy = candy17.copy()
#Remove data with null values
candy17_copy = candy17_copy.dropna()
candy17_copy.isnull().sum()
```

```python
candy17_copy.shape
```

As can be seen above, every row had at least one missing value resulting in the removal of all data. This is not ideal and would not be the preferred method when working with this data.

### 1.5 Fill in missing data

```python
#Replace null values
candy17 = candy17.fillna("No Response")
candy17.head()
```

### 2. Replace Values

```python
#Replace united states country value
candy17['Q4: COUNTRY'] = candy17['Q4: COUNTRY'].replace({'United States' : 'USA', 'Usa':'USA','US':'USA',
                                                        'usa':'USA','us':'USA','Murica':'USA'})
candy17.head(10)
```

### 3. Create hierarchical index

```python
#Set hierarchical index
candy17_hier = candy17.set_index(['Internal ID','Q2: GENDER'])
candy17_hier
```

### 4. Reshape

```python
#Pivot columns into rows with stack
stacked = candy17_hier.stack()
stacked
```

```python
#Return data to original state
stacked.unstack()
```

### 5. Grouping with Index Levels

```python
#Group hierarchical indexed df by gender responses
candy17_hier.groupby(level='Q2: GENDER').count()
```

### 6. Split/Apply/Combine

```python
#Replace non-numeric values with Nan
candy17['Q3: AGE'] = candy17['Q3: AGE'].apply(lambda x: pd.to_numeric(x, errors='coerce') )
#Replace null values with zero
candy17['Q3: AGE'] = candy17['Q3: AGE'].fillna(0)
#Check for remaining null values
candy17['Q3: AGE'].isnull().sum()
```

```python
#Create function to find 10 oldest respondents
def you_are_old (candy17, n=10, column='Q3: AGE'):
    return candy17.sort_values(by=column)[-n:]

candy17.groupby('Q2: GENDER').apply(you_are_old)
```
